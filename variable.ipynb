```python
class make_car():
    def __init__(self):
        self.door_color="red"
        self.front_color="pink"
        self.back_color="blue"
    def price(self,x):
        print(f'{x} price is quite cheap')
    
c1=make_car()
print(c1.door_color)
print(c1.front_color)
print(c1.back_color)
c1.price(10)
        
```

```python
import psutil
import os
info = psutil.virtual_memory()
print (u"記憶體使用",psutil.Process(os.getpid()).memory_info().rss)
print (u"總記憶體：",info.total)
print (u"記憶體佔比：",info.percent)
print (u"cpu個數：",psutil.cpu_count())
```
